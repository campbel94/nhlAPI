Franchises
GET https://statsapi.web.nhl.com/api/v1/franchises Returns a list of franchises
GET https://statsapi.web.nhl.com/api/v1/franchises/ID Gets information on a specific franchise

Provides:
- Franchise ID
- First season
- Last season
- Most recent team ID
- Team name
- Location
- Link
