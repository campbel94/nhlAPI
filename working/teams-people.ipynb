**Modifiers**
- `?expand=team.roster` Shows roster of active players for the specified team
- `?expand=person.names` Same as above, but gives less info.
- `?expand=team.schedule.next` Returns details of the upcoming game for a team
- `?expand=team.schedule.previous` Same as above but for the last game played
- `?expand=team.stats` Returns the teams stats for the season
- `?expand=team.roster&season=20142015` Adding the season identifier shows the roster for that season
- `?teamId=4,5,29` Can string team id together to get multiple teams
- `?stats=statsSingleSeasonPlayoffs` Specify which stats to get. Not fully sure all of the values

Create on table containing each teams roster for each NHL season.

```python
seasons = []
team_ids = []
team_names = []
player_ids = []
player_names = []

year1=1917
year2=1918

while year2 < 2023:
    try:
        teams_url = 'https://statsapi.web.nhl.com/api/v1/teams?expand=team.roster&season={}{}'.format(year1,year2)
        res=requests.get(teams_url)
        for i in res.json()['teams']:
            team_id = i['id']
            team_name = i['franchise']['teamName']
            for j in i['roster']['roster']:
                seasons.append(f'{year1}-{year2}')
                team_ids.append(team_id)
                team_names.append(team_name)
                player_ids.append(j['person']['id'])
                player_names.append(j['person']['fullName'])
    except:
        pass
    year1+=1
    year2+=1
# print(team_ids, team_names, player_ids, player_names)
```

```python
player_ids = []
player_names = []
player_birthdays = []
player_birthcities = []
player_birthcountries = []
player_nationalities = []
player_heights = []
player_weights = []
player_handedness = []
player_positions = []


for i in rosters_df['player_ids'].unique():
    players_url = 'https://statsapi.web.nhl.com/api/v1/people/{}'.format(i)
    res = requests.get(players_url).json()
    try:
        player_ids.append(res['people'][0]['id'])
    except:
        player_ids.append('NA')
    try:
        player_names.append(res['people'][0]['fullName'])
    except:
        player_names.append('NA')
    try:
        player_birthdays.append(res['people'][0]['birthDate'])
    except:
        player_birthdays.append('NA')
    try:
        player_birthcities.append(res['people'][0]['birthCity'])
    except:
        player_birthcities.append('NA')
    try:
        player_birthcountries.append(res['people'][0]['birthCountry'])
    except:
        player_birthcountries.append('NA')
    try:
        player_nationalities.append(res['people'][0]['nationality'])
    except:
        player_nationalities.append('NA')
    try:
        player_heights.append(res['people'][0]['height'])
    except:
        player_heights.append('NA')
    try:
        player_weights.append(res['people'][0]['weight'])
    except:
        player_weights.append('NA')
    try:
        player_handedness.append(res['people'][0]['shootsCatches'])
    except:
        player_handedness.append('NA')
    try:
        player_positions.append(res['people'][0]['primaryPosition']['name'])
    except:
        player_positions.append('NA')
```

```python
players_df = pd.DataFrame({
    'player_ids': player_ids,
    'player_names': player_names,
    'player_birthdays': player_birthdays,
    'player_birthcities': player_birthcities,
    'player_birthcountries': player_birthcountries,
    'player_nationalities': player_nationalities,
    'player_heights': player_heights,
    'player_weights': player_weights,
    'player_handedness': player_handedness,
    'player_positions': player_positions
})
players_df
```
